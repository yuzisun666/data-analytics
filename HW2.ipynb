Some colors are very similar to each other and it may be hard to understand what property they represent. We are going to focus on interactive visualization in future and thus I am just going to share a few lines of code 
that can help us read such graphs:


1.   We will need to load interactive visualization library **plotly** (Note that some functions' codes for the same or similar graph may slightly change - different people created different Python/R libraries and they decided to NOT be consistent. We will assign px alias for plotly
2.   We will use line(...) function in plotly library to create an interactive graph. 

3.   Interactive in the following sense: when graph is created we will hover mouse on any line and marker and we will see what property it represents and we will not need to guess what many shades of the same color mean. 





```python
# 27 Import plotly library: run the line code below
# The plotly.express module (usually imported as px) contains functions that can create entire 
# figures at once, and is referred to as Plotly Express or PX. Plotly Express is the easy-to-use, 
# high-level interface to Plotly, which operates on a variety of types of data and produces easy-to-style figures.
# Remark: this question does not require you to do anything besides running this line of code
import plotly.express as px
```

```python
# 28. Let's run this line of code to create a line graph that is interactive: 
# hover an each trajectiry with a mouse and it will show you property name 
# and rent per sq ft 
# Remark: you do not need to craete code to answer this question, you just need to run it and 
# hover on different trajectories and experience what is meant by interactive visualization
# feel free to play with different zip codes. 
px.line(rents.loc[rents.PROPERTY_ZIPCODE==90046,],
        x="YEAR_MONTH", y= "RENT_PER_SQFT_ACTUAL",
        color='PROPERTY_NAME', height=600, width=1000,markers=True)
```

**Remark**: when a graph has many trajectories that are far from each other, each trajectory's local patterns get squashed and it is hard to see them. By patterns I mean increasing/decreasing trend, seasonality, cyclicality of data, etc. Plotting only one property's data over time(or properties whose values are on the same scale), can help us better see what's going on with data over time.
