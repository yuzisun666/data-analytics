```python
import pandas as pd
add = pd.read_csv('https://raw.githubusercontent.com/robertasgabrys/DSO559/main/AdditionalInformation.csv', usecols= [0,1,2,3,4,5], skiprows=[0])
add.head()
```

```python
add.columns= ['Store', "Lay's", 'Popchips', 'Pringles', 'Sun Chips', 'Tostitos']
add.head()
```

```python
add.set_index('Store', inplace=True)
add.head()
```

```python
add = add.unstack().reset_index()
add.head()
```

```python
add.columns= ['Chip brand', 'Store', 'Wholesale price']
add.head()
```

```python
import pandas as pd
add1 = pd.read_csv('https://raw.githubusercontent.com/robertasgabrys/DSO559/main/AdditionalInformation.csv', usecols= [8,9], skiprows=[0])
add1.head()
```

```python
add1.columns=['Store', 'Percentage']
add1.head()
```

```python
add = pd.merge(add,add1,left_on=['Store'],right_on=['Store'], how='inner')
add.head()
```

```python
add = pd.merge(add,chip,left_on=['Store','Chip brand'],right_on=['Store','Chip brand'], how='inner')
add.head()
```

```python
#total wholesale price = wholesale price of 1 bag of chips (from Table 1) * Total number of bags sold
add['total wholesale price']  = add['Wholesale price'] * add['Number of bags sold']
```

```python
add.groupby(['Store']).agg({'total wholesale price':'sum'}).min()
```

```python
#total sales revenue per storeï¼Ÿ take min()
#sales revenue per 1 bag = wholesale price of 1 bag* (1+markup percentage/100)
add['sales revenue per bag']  = add['Wholesale price'] * (add['Percentage']/100+1)
```

```python
add['total rev per store'] = add['sales revenue per bag'] * add['Number of bags sold']
```

```python
add.groupby(['Store']).agg({'total rev per store':'sum'}).max()
```

```python
#total profit per store. take max
#profit = retail price - wholesale price
add['profit'] = add['sales revenue per bag'] - add['Wholesale price']
```

```python
add['total profit'] = add['profit'] * add['Number of bags sold']
```

```python
add.groupby(['Store']).agg({'total profit':'sum'}).min()
```
